# Familiarity Enjoyment Rating box plot

```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import os

inDir = r'C:\Users\napat\Documents\GitHub\BCC_2024\behavioralRatings.mat'

# Load the .mat file using h5py for MATLAB v7.3 files
with h5py.File(inDir, 'r') as f:
    behavioral_ratings = np.array(f['behavioralRatings']) 

behavioral_ratings = behavioral_ratings.transpose()

f_size = 16
yl = [0, 9]

flierprops = dict(marker='+', color='red', markersize=10)

plt.figure(figsize=(10, 8)) 

# Familiarity ratings (Q1)
plt.subplot(2, 1, 1)
plt.boxplot(behavioral_ratings[:, :, 0], flierprops=flierprops)
plt.axhline(y=0, color='black')  # Adding y-axis line
plt.axvline(x=0.5, color='black')  # Adding x-axis line at the start of the plot
plt.text(10.5, 10.5, 'Q1: Familiarity', fontsize=f_size, horizontalalignment='right')
plt.ylabel('Rating', fontsize=f_size)
plt.ylim(yl)
plt.yticks([1, 5, 9])
plt.box(False)

# Enjoyment ratings (Q2)
plt.subplot(2, 1, 2)
plt.boxplot(behavioral_ratings[:, :, 1], flierprops=flierprops)
plt.axhline(y=0, color='black') 
plt.axvline(x=0.5, color='black') 
plt.text(10.5, 10.5, 'Q2: Enjoyment', fontsize=f_size, horizontalalignment='right')
plt.xlabel('Song Number', fontsize=f_size)  
plt.ylabel('Rating', fontsize=f_size)
plt.ylim(yl)
plt.yticks([1, 5, 9])
plt.box(False)

# Spacing between 2 graph
plt.subplots_adjust(hspace=0.5)  
output_dir = r'C:\Users\napat\Documents\GitHub\BCC_2024\Code\code_testing\Graph\Rating'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

output_file = os.path.join(output_dir, f'Familiarity_Enjoyment.png')

plt.savefig(output_file, bbox_inches='tight')
plt.show()
```
